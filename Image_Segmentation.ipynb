```python
# Importing required libraries
from skimage.segmentation import slic
from skimage.data import astronaut
from skimage.color import label2rgb

# Setting the plot size as 15, 15
plt.figure(figsize=(15,15))

# Sample Image of scikit-image package
astronaut = astronaut()

# Applying Simple Linear Iterative
# Clustering on the image
# - 50 segments & compactness = 10
astronaut_segments = slic(astronaut,
						n_segments=1,
						compactness=10)
plt.subplot(1,2,1)

# Plotting the original image
plt.imshow(astronaut)
plt.subplot(1,2,2)

# Converts a label image into
# an RGB color image for visualizing
# the labeled regions.
plt.imshow(label2rgb(astronaut_segments,
					astronaut,
					kind = 'avg'))
```

```python
# import the necessary packages
from skimage.segmentation import slic
from skimage import color
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

image = mpimg.imread('./Datasets/Images/konosuba.jpg')
numSegments =350
segments = slic(image, n_segments = numSegments, sigma = 5, compactness=10)

plt.figure(figsize=(30,30))
plt.subplot(1,2,1)
plt.imshow(image)
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(color.label2rgb(segments, image, kind = 'avg'))
plt.show()
```

```python
def replaceRand(arr):
    rep = np.random.randint(1000,size=len(np.unique(arr)))
    dictr = dict(zip(np.unique(arr), rep))
    for i in range(arr.shape[0]):
        for j in range(arr.shape[1]):
            arr[i][j] = dictr[arr[i][j]]
    return arr
mseg = replaceRand(segments)
plt.figure(figsize=(30,30))
plt.imshow(mseg)
# np.unique(segments)
```

```python
from skimage.segmentation import slic
import skimage
from skimage import color
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
from skimage.data import astronaut

# image = mpimg.imread('./Datasets/Images/konosuba.jpg')
image = astronaut()
print(image.shape)
numSegments = 30
segments = slic(image, n_segments = numSegments, sigma = 5, compactness=10)

plt.figure(figsize=(25,25))
plt.subplot(1,2,1)
plt.imshow(image)
plt.axis("off")

rgb_image = color.label2rgb(
  segments, colors=np.random.random((50, 3),))
plt.subplot(1,2,2)
plt.imshow(rgb_image)

plt.show()
```
